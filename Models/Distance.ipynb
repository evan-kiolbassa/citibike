```python
import pandas as pd
import numpy as np
import pickle
from sklearn.neighbors import DistanceMetric
from math import radians
```

```python
stations = pd.read_csv('../data/stations_cleaned.csv.gz')
```

```python
ride_stations = pd.read_csv('../data/riders_cleaned.csv.gz', usecols = ['start station id', 'start station name', 'start station latitude', 'start station longitude'])
```

```python
ride_stations
```

```python
# Method 1
ride_stations_df = ride_stations.groupby('start station id')[['start station latitude', 'start station longitude']].mean().reset_index()
```

```python
# Method 2
ride_stations_df = ride_stations.drop_duplicates(subset = ['start station id'])
```

```python
ride_stations_df
```

```python
ride_stations_df=ride_stations_df.rename(columns = {'start station id': 'dock_id'})
```

```python
ride_stations_df = ride_stations_df.reset_index()
```

```python
ride_stations_df = ride_stations_df.drop('index', axis = 1)
```

```python
ride_stations_df
```

```python
# Convert the Lat/Long degress in Radians
ride_stations_df['lat'] = np.radians(ride_stations_df['start station latitude'])
ride_stations_df['long'] = np.radians(ride_stations_df['start station longitude'])
```

```python
# Scipy get_metrics()
dist = DistanceMetric.get_metric('manhattan')
```

```python
ride_stations_df[['lat','long']].to_numpy()
```

```python
dist.pairwise(ride_stations_df [['lat','long']].to_numpy())*6373
```

```python
ride_stations_df
```

```python
Distance3 = pd.DataFrame(dist.pairwise(ride_stations_df[['lat','long']].to_numpy())*6373, columns=ride_stations_df.dock_id.unique(), index=ride_stations_df.dock_id.unique())
```

```python
pd.set_option('display.max_columns', None)
```

```python
pd.set_option('display.max_rows', None)
```

```python
Distance3 = Distance3.sort_index()
```

```python
Distance3.to_csv("../data/distance3.csv")
```
