```python
import pandas as pd
import numpy as np
import pickle
```

```python
stations = pd.read_csv('../data/stations_cleaned.csv.gz')
```

```python
stations
```

```python
station_dict={}
for station in stations.dock_id.unique().tolist():
    loc1=(stations[stations.dock_id==station]._lat.values,stations[stations.dock_id==station]._long.values)
    others=[x for x in stations.dock_id.unique().tolist() if x != station]
    dist_to_others={}
    for othr_station in others:
        loc2=(stations[stations.dock_id==othr_station]._lat.values,stations[stations.dock_id==othr_station]._long.values)
        distance=(abs(loc1[0]-loc2[0])+abs(loc1[1]-loc2[1]))[0]
        dist_to_others[othr_station]=distance
    dist_to_others=sorted(dist_to_others.items(), key = lambda kv:(kv[1], kv[0])) 
    station_dict[station]=dist_to_others


output = open('station_dict.pkl', 'wb')
pickle.dump(station_dict, output, -1)
output.close()
```
