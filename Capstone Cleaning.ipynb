```python
import glob
import pandas as pd
import numpy as np
import regex as re
import missingno as msno
        
def path_to_dict(path):
    '''
    Takes a file path in as input and return a dictionary of dataframes
    where the filename without the .csv is the key for the corresponding 
    dataframe.
    '''
    files = glob.glob(path + "*.csv")
    path = re.sub('\W+', " ", path)
    dataframe_collection = {} 
    for file in files:
        key = re.sub('\W+', " ", file)
        key = key.replace(path,'')
        key = key.replace('csv','')
        dataframe_collection[key] = pd.read_csv(file)
    return dataframe_collection
```

```python
trip_data = path_to_dict(r'F:\Capstone Data\\')
```

```python
for k,v in trip_data.items():
    k = re.findall('\d+',k)
```

```python
key_list = []
df_list = []
for k,v in trip_data.items():
    v.columns = pd.Series(v.columns).str.lower()
    v.columns = pd.Series(v.columns).str.strip()
    v.columns = pd.Series(v.columns).str.replace(" ", "")
    v.loc[:, ['starttime', 'stoptime']] = pd.to_datetime(v.loc[:, ['starttime', 'stoptime']].stack(), 
    format='%Y-%m-%d %H:%M:%S', infer_datetime_format = True).unstack()
    key_list.append(k)
    df_list.append(v)
```

```python
key_list = pd.Series(key_list).str.replace('\D+', '', regex = True)
key_list = key_list.str.slice(stop=4)
```

```python
year_list = ['2013', '2014', '2015','2016','2017','2018','2019','2020']
```

```python
dict_indexes = {}
for year in year_list:
    dict_indexes[year] = key_list[key_list == year].index.tolist()
```

```python
df_list = pd.Series(df_list)
```

```python
quik_2013 = pd.concat(df_list[dict_indexes['2013']].tolist())
quik_2014 = pd.concat(df_list[dict_indexes['2014']].tolist())
#quik_2015 = pd.concat(df_list[dict_indexes['2015']].tolist())
quik_2016 = pd.concat(df_list[dict_indexes['2016']].tolist())
quik_2017 = pd.concat(df_list[dict_indexes['2017']].tolist())
quik_2018 = pd.concat(df_list[dict_indexes['2018']].tolist())
quik_2019 = pd.concat(df_list[dict_indexes['2019']].tolist())
quik_2020 = pd.concat(df_list[dict_indexes['2020']].tolist())
```
